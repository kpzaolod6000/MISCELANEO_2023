```python
from itertools import islice
from tqdm.notebook import tqdm
from re import sub
num_lines = sum(1 for line in open("wiki_tfidf_terms.csv"))
with open("wiki_tfidf_terms.csv") as file:
    dict_idf = {}
    with tqdm(total=num_lines) as pbar:
        for i, line in tqdm(islice(enumerate(file), 1, None)):
            try: 
                cells = line.split(",")
                idf = float(sub("[^0-9.]", "", cells[3]))
                dict_idf[cells[0]] = idf
            except: 
                print("Error on: " + line)
            finally:
                pbar.update(1)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from numpy import array, log

def myf3(x):
  vectorizer = CountVectorizer()
  names = x
  tf = vectorizer.fit_transform([ names])
  # for index, row in df.iterrows():
  #   if type(row['abstract']) == str:
  #    names.append(row.get('abstract'))
    #  tf = vectorizer.fit_transform([x.lower() for x in names])
  tf = tf.toarray()
  tf = log(tf + 1)

  tfidf = tf.copy()
  words = array(vectorizer.get_feature_names())
  for k in tqdm(dict_idf.keys()):
      if k in words:
          tfidf[:, words == k] = tfidf[:, words == k] * dict_idf[k]
      pbar.update(1)
  for j in range(tfidf.shape[0]):
      # print("Keywords of article", str(j+1))
      keywords = words[tfidf[j, :].argsort()[-5:][::-1]]
      # print(keywords)
      return keywords
```

```python
for index, row in df.iterrows():
  #  print(row["Bert_synonym"])
  if type(row['Abstracts']) == str: 
    x = row['Abstracts']
    cosine = myf3(x)
    # print(cosine)
    df.iloc[index, df.columns.get_loc("keywords_tfidf")] = myfunc1(cosine)
```

```python
# Program to measure the similarity between 
# two sentences using cosine similarity.

def func3(X,Y):
  # X = input("Enter first string: ").lower()
  # Y = input("Enter second string: ").lower()    
  # tokenization

  X_list = word_tokenize(X) 
  Y_list = word_tokenize(Y)
  # print(X_list)
  # print(Y_list)
  # sw contains the list of stopwords
  sw = stopwords.words('english') 
  l1 =[];l2 =[]
    
  # remove stop words from the string
  X_set = {w for w in X_list if not w in sw} 
  Y_set = {w for w in Y_list if not w in sw}
  # print(X_set)
  # print(Y_set)
  # print("****")
  # form a set containing keywords of both strings 
  rvector = X_set.union(Y_set) 
 
  for w in rvector:
      if w in X_set: 
        l1.append(1) # create a vector
      else:
        l1.append(0)
      if w in Y_set:
        l2.append(1)
        #  print(l2)
      else:
        l2.append(0)
  c = 0
  # cosine formula 
  for i in range(len(rvector)):
          c+= l1[i]*l2[i]
  cosine = c / float((sum(l1)*sum(l2))**0.5)
  return cosine
```

```python

for index, row in df.iterrows():
  if type(row['Keywords']) == str: 
   x = row['Keywords']
   y= row['keywords_Bert']
   cosine = func3(x,y)
   df.iloc[index, df.columns.get_loc("Bert_similarity")] = cosine
```
